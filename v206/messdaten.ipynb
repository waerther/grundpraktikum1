Gemessene Daten:

```python
import pandas as pd
messdaten = pd.read_excel("messdaten.xlsx")
daten = messdaten.iloc[:,[0, 2, 3, 4, 6, 7, 10]] 
print(daten.to_latex(index = False, column_format= "c c c c c c", na_rep = "-", float_format="%.2f"))
```

berechnete Werte:

```python
import numpy as np
werte = messdaten.iloc[:,[1, 4, 5, 8, 9, 11]] 
werte.iloc[:,[5]] = werte.iloc[:,[5]].applymap("{0:.4f}".format)
werte.iloc[:,[0,1,2,3,4]] = werte.iloc[:,[0,1,2,3,4]].applymap("{0:.2f}".format)
print(werte.to_latex(index = False, column_format= "c c c c c c", na_rep = "-")) #, float_format= "%.2f"
```
